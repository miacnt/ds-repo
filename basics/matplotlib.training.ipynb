```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel("time")
plt.ylabel("some value")
```
